```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Box Plots**

```python
sns.boxplot(x='sex', y='combined_respondent_prestige', data=df)
```

```python
sns.boxplot(x='marital_status', y='combined_respondent_prestige', data=df)
```

**Grouped Box Plots**

```python
df.boxplot(column = 'combined_respondent_prestige',by = 'sex')
df.boxplot(column = 'combined_respondent_prestige',by = 'marital_status')
```

**Violin Plots**

```python
sns.violinplot(x='sex', y='combined_respondent_prestige', data=df)
```

```python
sns.violinplot(x='marital_status', y='combined_respondent_prestige', data=df)
```
