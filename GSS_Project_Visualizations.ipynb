Research Question: How does the distribution of occupational prestige vary among different levels of marriage happiness and sexes in the married population?

```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/My Drive/4th Year/cleaned_gss.csv', index_col=0)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.kdeplot(data=df,x='combined_respondent_prestige',hue='marriage_happiness')
plt.title('Distribution of Combined Respondent Prestige Across Marriage Happiness Levels')
plt.xlabel('Combined Occupational Prestige')
plt.ylabel('Density')
plt.legend(title='Marriage Happiness', labels=['Very Happy', 'Pretty Happy', 'Not Too Happy'])
```

```python
sns.kdeplot(data=df,x='combined_respondent_prestige',hue='sex')
plt.title('Distribution of Combined Respondent Prestige Across Sexes')
plt.xlabel('Combined Occupational Prestige')
plt.ylabel('Density')
plt.legend(title='Sex', labels=['Female', 'Male'])
```

```python
df['marriageXsex']=df['marriage_happiness']+', '+df['sex']
sns.kdeplot(data=df,x='combined_respondent_prestige',hue='marriageXsex')

plt.title('Distribution of Combined Respondent Prestige Across Marriage Happiness Levels and Sexes')
plt.xlabel('Combined Occupational Prestige')
plt.ylabel('Density')
plt.legend(title='Marriage Happiness and Sex ', labels=['Very Happy, Female', 'Very Happy, Male','Pretty Happy, Female','Pretty Happy, Male', 'Not Too Happy, Female', 'Not Too Happy, Male'], fontsize='small')
```
